```python
class Student:
    def __init__(self, matrix) -> None:
        self.matrix = matrix

class Portal(Student):
    def __init__(self, matrix) -> None:
        super().__init__(matrix)
        self.matrix[0].append('Grade')

    def assign_grade(self, sapid):
        for i in range(1, len(self.matrix), 1):
            # # self.matrix[i].append("Math")
            if self.matrix[i][1] == sapid:
                percentage = self.matrix[i][3]/25 * 100
                if percentage > 80:
                    self.matrix[i].append('A')
                elif percentage < 80 and percentage > 70:
                    self.matrix[i].append('B')
                else:
                    self.matrix[i].append('C')
                return

    def display_result(self, rnum):
        for i in range(1, len(self.matrix), 1):
            if self.matrix[i][1] == rnum:
                print(self.matrix[i])



student = [
    ['Name', 'SapId', 'Subject', 'Marks'],
    ['Hasham', '70120545', 'Math', 14],
    ['Ali', '70121929', 'Biology', 25]
    ]

# h = ['hello']

std = Portal(student)

std.assign_grade('70121929')

std.display_result('70121929')




# student.append("Hasham")

# print(student)
```
